```python
import statsmodels.api as sm
```

```python
x = df[['TV']]
y = df['Sales']

x = sm.add_constant(x)

model = sm.OLS(y, x).fit()

print(model.summary())
```

```python
sns.residplot(x=model.fittedvalues, y=model.resid)

plt.show()
```

```python
sm.qqplot(model.resid, line='45')
plt.show()
```
